# Question 3

```python
df = pd.read_csv('menu.csv')


# Select only numeric columns for correlation analysis
numeric_columns = df.select_dtypes(include=['int64', 'float64'])

# Calculate correlation matrix
correlation_matrix = numeric_columns.corr()

# Plot correlation matrix
plt.figure(figsize=(12, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python


# Get unique categories
categories = df['Category'].unique()

# Set figure size
plt.figure(figsize=(12, 5 * len(categories)))  # Adjust figure size based on the number of categories

# Iterate over each category and plot a bar graph
for i, category in enumerate(categories, start=1):
    # Filter DataFrame for the current category
    category_df = menu_df[df['Category'] == category]
    
    # Sort items by calories in descending order
    category_df = category_df.sort_values(by='Calories', ascending=False)
    
    # Truncate item names based on category
    if category == 'Coffee & Tea':
        category_df['Item'] = category_df['Item'].apply(lambda x: ' '.join(x.split()[:1]))
    else:
        category_df['Item'] = category_df['Item'].apply(lambda x: ' '.join(x.split()[:4]))
    
    # Create subplot for the current category
    plt.subplot(len(categories), 1, i)
    
    # Plot bar graph
    plt.barh(category_df['Item'], category_df['Calories'], color='skyblue')
    
    # Set labels and title
    plt.xlabel('Calories')
    plt.ylabel('Item')
    plt.title(f'Calories by Item in {category}')
    
    # Rotate item labels for better readability
    plt.xticks(rotation=45, ha='right')
    
    # Adjust layout
    plt.tight_layout()

plt.show()
```
